```python
def entropy(col):                                       # Calculating Entropy 
    types = np.unique(col, return_counts = True)
    entropy = 0.0
    
    N = col.shape[0]
    for ix in types[1]:
        probability = ix/N
        entropy += (-1.0*probability*np.log2(probability))
        
    return entropy
```

```python
def split_on(system , fkey ):
    klasses = np.unique(system[fkey], return_counts=True)
    key , val = list() , list()
    for Klass in klasses[0]:
        data = pd.DataFrame([] , columns=system.columns)
        for ix in range (system.shape[0]):
             if Klass == system[fkey].loc[ix]:
                    data=data.append(system.loc[ix])
        data =data.reset_index(drop = True)
        val.append(data)
        key.append(Klass)
        
    return key ,val
```

```python
def information_gain(system , fkey):
    key , val = split_on(system , fkey)
    info_gain = entropy(system['class'])
    for i in range (len(key)):
        value=val[i].shape[0]/system.shape[0]
        info_gain -= value*entropy(val[i]['class'])
    return info_gain
```

```python
class Generic_decision_Trees():
    def __init__ (self , max_depth = 8 , depth =0):
        self.target = None
        self.fkey = None
        self.node = None
        self.child = list()
        self.max_depth = max_depth
        self.depth = depth
        
    def train(self , train):
        ds  =  ['cap-shape','cap-surface','cap-color','bruises','odor','gill-spacing','gill-size','gill-color','stalk-shape',
                'stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring',
                'stalk-color-below-ring','ring-number','ring-type','spore-print-color','population','habitat']
        ig=[]
        for info in ds:
            ig.append(information_gain(train , info))
        print(ig)
        self.fkey = ds[np.argmax(ig)]
        print(self.fkey)
        self.node , data = split_on(train , self.fkey)
        
        if self.depth >= self.max_depth:
            ans , count = np.unique(train['class'] ,return_counts=True)
            cmax = ans[count.argmax()]
            self.target = cmax
            return
        if len(data)==1:
            ans , count = np.unique(data[0]['class'] ,return_counts=True)
            cmax = ans[count.argmax()]
            self.target = cmax
            return 
        
        for i in range (len(self.node)):
            nc = Generic_decision_Trees(depth=self.depth+1,max_depth=self.max_depth)
            nc.train(data[i])
            self.child.append(nc)
        
        # Appending the values of target
        ans , count = np.unique(train['class'] ,return_counts=True)
        cmax = ans[count.argmax()]
        self.target = cmax
        
    def predict(self , test):
        #print(self.node,test[self.fkey])
        if len(self.child)==0:
            return self.target
        if test[self.fkey] in self.node:
            return self.child[self.node.index(test[self.fkey])].predict(test)
        else :
            return self.target
        
        
        
```

```python
class Random_forest():
    
    def __init__ (self , n_estimators=5  , max_depth = 5 , batch_size = 0.8):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.batch_size = batch_size
        self.branch = list()
    
    def fit(self ,X_train):
        for i in range (self.n_estimators):
            x=X_train
            split = int(self.batch_size*x.shape[0])
            features = x.columns
            x = x.values
            np.random.shuffle(x)
            x = x[:split , :]
            x = pd.DataFrame(x , columns=features)
            dt = Generic_decision_Trees(max_depth=self.max_depth)
            dt.train(x)
            self.branch.append(dt)
    def score(self ,X_test):
        random_forest = list()
        for j in range (X_test.shape[0]):
            pred = list()
            for i in range (len(self.branch)):
                pred.append(self.branch[i].predict(X_test.loc[j]))
            unique = np.unique(pred , return_counts=True)
            
            random_forest.append(unique[0][np.argmax(unique[1])])
            
        return np.sum(random_forest==X_test['class'])/X_test.shape[0]
```

```python
rf = Random_forest(n_estimators=5 , max_depth=5)
```
